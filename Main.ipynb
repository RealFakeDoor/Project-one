```python
FedFund_df = pd.read_csv('./CSV_DATA/FEDFUND_DF.csv')

FedFund_df = FedFund_df.rename(columns={'date': 'Date'})

# Convert 'Date' column to datetime format
FedFund_df['Date'] = pd.to_datetime(FedFund_df['Date'])

# Define start and end dates
start_date = ['2003-01-01',           '2003-01-01',          '2006-06-06']
end_date =   ['2010-01-01',           '2006-06-06',          '2010-01-01']

    # Filter DataFrame between start and end dates
for i in range(0,len(start_date)):
    FedFund_TS = FedFund_df[(FedFund_df['Date'] >= start_date[i]) & (FedFund_df['Date'] <= end_date[i])]
    FedFund_TS.plot('Date','Federal Fund Rate Target')
    plt.title(f'Federal Fund Rate Target( {start_date[i]} - {end_date[i]} )')
```

```python
Correlation_matrix_rising = df_Rising_FEDFUND.corr()
Correlation_matrix_lowering = df_Lowering_FEDFUND.corr()
```

```python
# Plotting heatmap
plt.figure(figsize=(10, 8))
plt.imshow(Correlation_matrix_rising, cmap='inferno', interpolation='nearest')
plt.colorbar(label='Correlation')
plt.title('Interest Rate Rising: Correlation Heatmap', fontsize=20)
plt.xticks(range(len(Correlation_matrix_rising.columns)), Correlation_matrix_rising.columns, rotation=90, fontsize=12)
plt.yticks(range(len(Correlation_matrix_rising.columns)), Correlation_matrix_rising.columns, fontsize=12)
plt.tight_layout()

# Save the figure
plt.savefig('./Saved_Figures/HEATMAP_InterestRates_Rising.png')

plt.show()
```

```python
# Plotting heatmap
plt.figure(figsize=(10, 8))
plt.imshow(Correlation_matrix_lowering, cmap='inferno', interpolation='nearest')
plt.colorbar(label='Correlation')
plt.title('Interest Rates Falling: Correlation Heatmap',fontsize=20)
plt.xticks(range(len(Correlation_matrix_lowering.columns)), Correlation_matrix_lowering.columns, rotation=90, fontsize=12)
plt.yticks(range(len(Correlation_matrix_lowering.columns)), Correlation_matrix_lowering.columns, fontsize=12)
plt.tight_layout()

# Save the figure
plt.savefig('./Saved_Figures/HEATMAP_InterestRates_Falling.png')

plt.show()
```

```python
# Calculate ROI for each column
roi_Lowering_df = df_Lowering_FEDFUND / df_Lowering_FEDFUND.iloc[0] - 1
# Calculate percentage change as ROI

# Plot the DataFrame with ROI values
roi_Lowering_df.plot(kind='line', figsize=(15, 10))

# Add horizontal line at y=0
plt.axhline(y=0, color='k', linestyle='--')


# Add ROI formula label
plt.text(0.08, 0.9, 'ROI = (Current Value / Initial Value) - 1', fontsize=13, transform=plt.gca().transAxes)

# Add labels, title and legend.
plt.xlabel('Date, d, (YYYY-MM-DD)', fontsize=20)
plt.ylabel('Return on Investment, ROI, (:)', fontsize=20)
plt.title('Interest Rate Falling: Sector ROI', fontsize=20)
plt.legend(fontsize=15)

# Save the figure
plt.savefig('./Saved_Figures/LINEGRAPH_ROI_InterestRates_Falling.png')

# Show the plot
plt.show()
```

```python
# Calculate summary statistics for each DataFrame
summary_rising = df_Rising_FEDFUND.describe()
summary_lowering = df_Lowering_FEDFUND.describe()

#Save Summary statistics for later.
summary_rising.to_csv(f'./OUTPUT_CSVs/Summary_statistics/Sectors_rising_rates_summary_statistics.csv', index=False)
summary_lowering.to_csv(f'./OUTPUT_CSVs/Summary_statistics/Sectors_falling_rates_summary_statistics.csv', index=False)

# Extract standard deviations
std_rising = summary_rising.loc['std']
std_lowering = summary_lowering.loc['std']

# Plot standard deviations in a bar graph
plt.figure(figsize=(10, 6))

# Position of bars on x-axis
bar_width = 0.35
index = range(len(std_rising))

# Plot bars for rising and lowering datasets with respective colors
light_blue = '#ADD8E6'  # Light blue color
light_red =  '#FFB6C1'   # Light red color

plt.bar(index, std_rising, bar_width, label='Rising Federal Fund Rate', color=light_blue)
plt.bar([i + bar_width for i in index], std_lowering, bar_width, label='Falling Federal Fund Rate', color=light_red)

# Adding labels and title
plt.xlabel('Sector', fontsize=15)
plt.ylabel('Standard Deviation, std', fontsize=15)
plt.title('Standard Deviation of Each Sector', fontsize=15)
plt.xticks([i + bar_width/2 for i in index], std_rising.index, rotation=25)  # Rotate x-axis labels
plt.legend()

# Add vertical lines between each sector x-value
for i in range(len(std_rising)):
    plt.axvline(x=i + bar_width*2, color='gray', linestyle='-', linewidth=1)

# Add faded grid lines
plt.grid(axis='y', alpha=0.5)

# Show plot
plt.tight_layout()

# Save the figure
plt.savefig('./Saved_Figures/BARGRAPHS_STD_Sectors.png')

plt.show()
```

```python
# Calculate volatility for each column
Sector_volatility_cut = Sector_Data_Total[['Financials', 'Healthcare', 'Consumer Staples','Energy']]

# Calculate volatility for each column
window = 26
volatility = Sector_volatility_cut.rolling(window=window).std().iloc[window - 1:] * np.sqrt(window)

# Plot volatility for each column
plt.figure(figsize=(10, 6))
for column in volatility.columns:
    plt.plot(Sector_volatility_cut.index[window - 1:], volatility[column], label=column)

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.title('Volatility of Each Column Over Time')

# Add legend
plt.legend(loc='center left', fontsize = 15)

# Set x-axis ticks
ticks = np.arange(0, len(Sector_volatility_cut.index[window - 1:]), len(Sector_volatility_cut.index[window - 1:]) // 6)  # Increase by 3
tick_labels = Sector_volatility_cut.index[window - 1:][ticks]
plt.xticks(ticks=ticks, labels=tick_labels, rotation=45)

# Add a vertical red line at the date 2006-06-07
plt.axvline(x='2006-06-07', color='red', linestyle='--', linewidth=1)

# Add text annotations for Interest Rate Environments
plt.text(0.05, 0.95, 'Interest Rate Rising Environment', transform=plt.gca().transAxes, fontsize=12, va='top', ha='left')
plt.text(0.78, 0.95, 'Interest Rate Falling Environment', transform=plt.gca().transAxes, fontsize=12, va='top', ha='right')

# Add faint gridlines
plt.grid(axis='y', alpha=0.3)

# Show plot
plt.tight_layout()

# Save the figure
plt.savefig('./Saved_Figures/Volatility_VS_TIME_SECTORS.png')

plt.show()
```
