**ROI Line Graphs:** <br>
ROI stands for Return on Investment it allows us to visualize how much money you're making (or losing) on an investment over time.<br>
 It's a metric that tells you how much profit you've made on an investment compared to its initial cost expressed as a percentage of the initial investment<br>
ROI is calculated by subtracting the initial cost of the investment from its final value, then dividing this new number by the cost of the investment.
.

A line graph of just prices shows the price movement over time, but it doesn't tell you how much money you've made or lost.  An ROI graph calculates and displays the return on investme<br>
<br>t. 
•	ROI graphs allow you to compare the performance of different investments on the same chart. This is difficult with line graphs based on price, as price scales might dif<br>
<br>fer.
•	By displaying ROI, you can easily see which investments are generating the best returns relative to their initial <br>
<br>cost.
•	ROI graphs can help visualize the risk-return relationship. Investments with higher potential returns often come with higher volatility (price fluctuati<br>
<br>ons). 
•	While price line graphs can show price trends, ROI graphs can highlight trends in your investment performance over a specified time<br>
 period


```python
# Calculate ROI for each column
roi_Total_df = df_Total_FEDFUND_Cycle / df_Total_FEDFUND_Cycle.iloc[0] - 1

# Calculate percentage change as ROI

# Plot the DataFrame with ROI values
roi_Total_df.plot(kind='line', figsize=(15, 8))

# Add horizontal line at y=0
plt.axhline(y=0, color='k', linestyle='--')

# Add vertical line at specific date
# Convert the target date to a datetime object

target_date = '2006-06-07'
# pd.to_datetime('2006-06-07')
print(target_date)

# plt.axvline(x='2006-06-07', color='r', linestyle='--') # This line of code doesn't seem to implement correctly the red vertical line is in the wrong location.
plt.axvline(x=df_Total_FEDFUND_Cycle.index.get_loc(target_date), color='r', linestyle='--')
# Add labels and title
plt.xlabel('Date')
plt.ylabel('Daily ROI')
plt.title('Sectors ROI Over Whole Interest Rate Cycle')

# Save the figure
plt.savefig('./Saved_Figures/LINEGRAPH_ROI_InterestRates_Total.png')

# Show the plot
plt.show()
```

A boxplot, also called a box and whisker plot, is a way to visually summarize a dataset and get a quick understanding of how the data is spread out.

•	A boxplot shows the middle 50% of the data. A line in the middle of the box indicates the median.
•	The boxplot also shows how spread out the data is. The whiskers (the lines extending from the box) represent the upper and lower quarters of the data.
•	The length of the whiskers tells you how much variation there is from the middle 50%.
•	Boxplots also show outliers.


```python
plt.figure(figsize=(10, 6))  
# Plot the distribution of values in a boxplot.
df_Rising_FEDFUND.boxplot(rot=90)

# Adding labels and title
plt.xlabel('Sector')
plt.ylabel('Sector Prices')
plt.title('Rising FEDFUNDRATE')

# Save the figure
plt.savefig('./Saved_Figures/BOXPLOTs_InterestRates_Rising.png')

# Show plot
plt.show()
```

By Plotting the two time periods as boxplots next to each other we are easily able to compare the changes in the spread of data under different conditions.

You can easily compare multiple datasets side-by-side using boxplots. 
This allows you to see how the data is distributed across different groups or conditions.

```python
plt.figure(figsize=(15, 8))

# Plot the distribution of values in boxplots for both time periods.

# Plot boxplots for rising Federal Fund Rate
plt.boxplot(df_Rising_FEDFUND.values, 
            positions=np.arange(len(df_Rising_FEDFUND.columns)), 
            widths=0.4, patch_artist=False, 
            boxprops=dict(color='blue'), 
            showfliers=False)

# Plot boxplots for lowering Federal Fund Rate
plt.boxplot(df_Lowering_FEDFUND.values, 
            positions=np.arange(len(df_Lowering_FEDFUND.columns)) + 0.5, 
            widths=0.4, 
            patch_artist=False, 
            boxprops=dict(color='red'), 
            showfliers=False)

# Adding labels and title
plt.xlabel('Sector', fontsize = 15)
plt.ylabel('Share Value, p, ($)', fontsize = 15)
plt.title('Distribution of Sector Values', fontsize = 15)

# Create custom legend
legend_labels = ['Falling Federal Fund Rate', 'Rising Federal Fund Rate']
legend_handles = [plt.Line2D([0], [0], color='red', lw=4),
                  plt.Line2D([0], [0], color='blue', lw=4)
                 ]
plt.legend(legend_handles, legend_labels, fontsize=20)


# Set x-axis ticks and labels
plt.xticks(np.arange(len(df_Lowering_FEDFUND.columns)) + 0.25, df_Lowering_FEDFUND.columns, rotation=10, fontsize=12)

# Add vertical lines between x-value columns
for i in range(len(df_Lowering_FEDFUND.columns) - 1):
    plt.axvline(x=i + 0.75, color='gray', linestyle='-', linewidth=0.8)

# Add faint horizontal gridlines
plt.grid(axis='y', alpha=0.5, linestyle = '--')

# Show plot
plt.tight_layout()

# Save the figure
plt.savefig('./Saved_Figures/BOXPLOTs_InterestRates_ALL.png')

plt.show()
```

```python
# THe following Code calculate the summary statistics of the data and plots them as Bar Graphs for an easy visual comparison.

# Calculate summary statistics for each DataFrame
summary_rising = df_Rising_FEDFUND.describe()
summary_lowering = df_Lowering_FEDFUND.describe()

#Save Summary statistics for later.
summary_rising.to_csv(f'./OUTPUT_CSVs/Summary_statistics/Sectors_rising_rates_summary_statistics.csv', index=False)
summary_lowering.to_csv(f'./OUTPUT_CSVs/Summary_statistics/Sectors_falling_rates_summary_statistics.csv', index=False)

# Extract standard deviations
std_rising = summary_rising.loc['std']
std_lowering = summary_lowering.loc['std']

# Plot standard deviations in a bar graph
plt.figure(figsize=(15, 8))

# Position of bars on x-axis
bar_width = 0.35
index = range(len(std_rising))

# Plot bars for rising and lowering datasets with respective colors
light_blue = '#ADD8E6'  # Light blue color
light_red =  '#FFB6C1'   # Light red color

plt.bar(index, std_rising, bar_width, label='Rising Federal Fund Rate', color=light_blue)
plt.bar([i + bar_width for i in index], std_lowering, bar_width, label='Falling Federal Fund Rate', color=light_red)

# Adding labels and title
plt.xlabel('Sector', fontsize=15)
plt.ylabel('Standard Deviation, std', fontsize=15)
plt.title('Standard Deviation of Each Sector', fontsize=15)
plt.xticks([i + bar_width/2 for i in index], std_rising.index, rotation=25)  # Rotate x-axis labels
plt.legend()

# Add vertical lines between each sector x-value
for i in range(len(std_rising)):
    plt.axvline(x=i + bar_width*2, color='gray', linestyle='-', linewidth=1)

# Add faded grid lines
plt.grid(axis='y', alpha=0.5)

# Show plot
plt.tight_layout()

# Save the figure
plt.savefig('./Saved_Figures/BARGRAPHS_STD_Sectors.png')

plt.show()
```

```python
# Calculate volatility for each column
Sector_volatility_cut = Sector_Data_Total[['Financials', 'Healthcare', 'Consumer Staples','Energy']]

# Calculate volatility for each chosen sector.
# Use a window of time that rolls with the calculation.
window = 26 #Weeks ~ (6 months)
volatility = Sector_volatility_cut.rolling(window=window).std().iloc[window - 1:] * np.sqrt(window)

# Plot volatility for each column
plt.figure(figsize=(15, 8))
for column in volatility.columns:
    plt.plot(Sector_volatility_cut.index[window - 1:], volatility[column], label=column)

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.title('Volatility of Each Sector VS Time')

# Add legend
plt.legend(loc='center left', fontsize = 15)

# Set x-axis ticks
ticks = np.arange(0, len(Sector_volatility_cut.index[window - 1:]), len(Sector_volatility_cut.index[window - 1:]) // 6)  # Increase by 3
tick_labels = Sector_volatility_cut.index[window - 1:][ticks]
plt.xticks(ticks=ticks, labels=tick_labels, rotation=45)

# Add a vertical red line at the date 2006-06-07
plt.axvline(x='2006-06-07', color='red', linestyle='--', linewidth=1)

# Add text annotations for Interest Rate Environments
plt.text(0.05, 0.95, 'Interest Rate Rising Environment', transform=plt.gca().transAxes, fontsize=12, va='top', ha='left')
plt.text(0.78, 0.95, 'Interest Rate Falling Environment', transform=plt.gca().transAxes, fontsize=12, va='top', ha='right')

# Label The length of the window of time to measure the volaility over.
plt.text(0.26, 0.70, f'Sample Length = {window} Weeks', transform=plt.gca().transAxes, fontsize=12, va='top', ha='right')

# Add faint gridlines
plt.grid(axis='y', alpha=0.3)

# Show plot
plt.tight_layout()

# Save the figure
plt.savefig('./Saved_Figures/Volatility_VS_TIME_SECTORS.png')

plt.show()
```
