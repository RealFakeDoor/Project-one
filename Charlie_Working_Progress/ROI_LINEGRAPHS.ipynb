```python
from pprint import pprint
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
```

```python
path = 'Total_FED_CYCLE.csv'

df_Total_FED_CYCLE = pd.read_csv(path)
```

```python
path = 'FED_FUND_RATE_TOTAL.csv'

FED_FUND_RATE_TOTAL = pd.read_csv(path)
```

```python
# Set 'Date' column as the index
df_Total_FED_CYCLE.set_index('Date', inplace=True)
```

```python
# Calculate ROI for each column
df_roi = roi_df = df_Total_FED_CYCLE / df_Total_FED_CYCLE.iloc[0] - 1
# Calculate percentage change as ROI

# Plot the DataFrame with ROI values
df_roi.plot(kind='line', figsize=(10, 6))

# Add horizontal line at y=0
plt.axhline(y=0, color='k', linestyle='--')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Daily ROI (%)')
plt.title('Sectors ROI Over the Entire Period')

# Show the plot
plt.show()
```
