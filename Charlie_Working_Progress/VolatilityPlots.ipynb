```python
from pprint import pprint
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
```

```python
path = 'FedFund_Rising.csv'
df_Rising_FEDFUND = pd.read_csv(path, index_col='Date')

path = 'FedFund_Lowering.csv'
df_Lowering_FEDFUND = pd.read_csv(path, index_col='Date')

path = 'Total_Fed_Cycle.csv'
df_Total_FEDFUND_Cycle = pd.read_csv(path, index_col='Date')

plt.figure(figsize=(10, 6))  

df_Lowering_FEDFUND.boxplot(rot=90, color = 'red')
df_Rising_FEDFUND.boxplot(rot=90)

# Adding labels and title
plt.xlabel('Sector')
plt.ylabel('Prices')
plt.title('Distribution of Sector Prices')
# Create custom legend
legend_labels = ['Lowering Federal Fund Rate', 'Rising Federal Fund Rate']
legend_handles = [plt.Line2D([0], [0], color='red', lw=4),
                  plt.Line2D([0], [0], color='blue', lw=4)]
plt.legend(legend_handles, legend_labels)

# Show plot
plt.show()
```
