```python
from pprint import pprint
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
```

```python
path = 'Total_Fed_Cycle.csv'

df = pd.read_csv(path)
```

```python
path = 'FedFund_Rising.csv'

df_Rising_FEDFUND = pd.read_csv(path)
```

```python
path = 'FedFund_Lowering.csv'

df_Lowering_FEDFUND = pd.read_csv(path)
```

```python
df_Rising_FEDFUND.drop(columns=['Date'], inplace=True)
df_Lowering_FEDFUND.drop(columns=['Date'], inplace=True)
```

```python
Correlation_matrix_rising = df_Rising_FEDFUND.corr()
Correlation_matrix_lowering = df_Lowering_FEDFUND.corr()
```

```python
# Plotting heatmap
plt.figure(figsize=(10, 8))
plt.imshow(Correlation_matrix_rising, cmap='inferno', interpolation='nearest')
plt.colorbar(label='Linear Correlation')
plt.title('Interest Rate Rising Correlation Heatmap')
plt.xticks(range(len(Correlation_matrix.columns)), Correlation_matrix.columns, rotation=90)
plt.yticks(range(len(Correlation_matrix.columns)), Correlation_matrix.columns)
plt.tight_layout()
plt.show()
```

```python
# Plotting heatmap
plt.figure(figsize=(10, 8))
plt.imshow(Correlation_matrix_lowering, cmap='inferno', interpolation='nearest')
plt.colorbar(label='Correlation')
plt.title('Interest Rate lowering Correlation Heatmap')
plt.xticks(range(len(Correlation_matrix.columns)), Correlation_matrix.columns, rotation=90)
plt.yticks(range(len(Correlation_matrix.columns)), Correlation_matrix.columns)
plt.tight_layout()
plt.show()
```

```python
# Calculate ROI for each column
roi_Lowering_df = df_Lowering_FEDFUND / df_Lowering_FEDFUND.iloc[0] - 1
# Calculate percentage change as ROI

# Plot the DataFrame with ROI values
roi_Lowering_df.plot(kind='line', figsize=(10, 6))

# Add horizontal line at y=0
plt.axhline(y=0, color='k', linestyle='--')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Daily ROI (%)')
plt.title('Sectors ROI Over Interest Rate Lowering Period')

# Show the plot
plt.show()
```

```python
# Calculate ROI for each column
roi_Rising_df = df_Rising_FEDFUND / df_Rising_FEDFUND.iloc[0] - 1
# Calculate percentage change as ROI

# Plot the DataFrame with ROI values
roi_Rising_df.plot(kind='line', figsize=(10, 6))

# Add horizontal line at y=0
plt.axhline(y=0, color='k', linestyle='--')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Daily ROI (%)')
plt.title('Sectors ROI Over Interest Rate Raising Period')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))  
df_Rising_FEDFUND.boxplot(rot=90)
# Adding labels and title
plt.xlabel('Sector')
plt.ylabel('Sector Prices')
plt.title('Distribution of Sector Prices PRE-2008 Stock Market Crash')

# Show plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))  
df_Lowering_FEDFUND.boxplot(rot=90)
# Adding labels and title
plt.xlabel('Sector')
plt.ylabel('Sector Prices')
plt.title('Distribution of Sector Prices PRE-2008 Stock Market Crash')

# Show plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))  
df_Lowering_FEDFUND.boxplot(rot=90, color = 'red')
df_Rising_FEDFUND.boxplot(rot=90)

# Adding labels and title
plt.xlabel('Sector')
plt.ylabel('Prices')
plt.title('Distribution of Sector Prices')
# Create custom legend
legend_labels = ['Lowering Federal Fund Rate', 'Rising Federal Fund Rate']
legend_handles = [plt.Line2D([0], [0], color='red', lw=4),
                  plt.Line2D([0], [0], color='blue', lw=4)]
plt.legend(legend_handles, legend_labels)

# Show plot
plt.show()
```
