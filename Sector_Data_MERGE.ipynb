```python
from pprint import pprint
import requests
import pandas as pd
```

```python
    Tickers_Dict = {
    'Energy':                     ['XLE'],
    'Materials':                  ['XLB'],
    'Industrials':                ['XLI'],
    'Consumer Discretionary':     ['XLY'],
    'Consumer Staples':           ['XLP'],
    'Health Care':                ['XLV'],
    'Financials':                 ['XLF'],
    'Information Technology':     ['XLK'],
    'Utilities':                  ['XLU'],
}
sector_list = list(Tickers_Dict.keys())
```

```python
# XLE Energy
path = './CSV_DATA/XLE_ENERGY_SELECT_PRE_2008.csv'
Energy_df = pd.read_csv(path)
ENERGY_DF = Energy_df[['Date','Close']]
ENERGY_DF = ENERGY_DF.rename(columns={'Close':'Energy Price'})
ENERGY_DF
```

```python
# XLB Materials
path = './CSV_DATA/XLB_MATERIALS_PRE_2008.csv'
Materials_df = pd.read_csv(path)
MATERIALS_DF = Materials_df[['Date','Close']]
MATERIALS_DF = MATERIALS_DF.rename(columns={'Close':'Materials Price'})
MATERIALS_DF
```

```python
# XLI Indistrials
path = './CSV_DATA/XLI_INDUSTRIALS_PRE_2008.csv'
Industrials_df = pd.read_csv(path)
INDUSTRIALS_DF = Industrials_df[['Date','Close']]
INDUSTRIALS_DF = INDUSTRIALS_DF.rename(columns={'Close':'Industrials Price'})
INDUSTRIALS_DF
```

```python
# XLY Consumer Discretionary
path = './CSV_DATA/XLY_CONSUMER_PRE_2008.csv'
Consumer_df = pd.read_csv(path)
CONSUMER_DF = Consumer_df[['Date', 'Close']]
CONSUMER_DF = CONSUMER_DF.rename(columns={'Close': 'Consumer Discretionary Price'})
CONSUMER_DF
```

```python
# XLP Consumer Staples
path = './CSV_DATA/XLP_CONSUMER_STAPLES_PRE_2008.csv'
Staples_df = pd.read_csv(path)
STAPLES_DF = Staples_df[['Date', 'Close']]
STAPLES_DF = STAPLES_DF.rename(columns={'Close': 'Consumer Staples Price'})
STAPLES_DF
```

```python
# XLV Healthcare
path = './CSV_DATA/XLV_HEALTHCARE_PRE_2008.csv'
Healthcare_df = pd.read_csv(path)
HEALTHCARE_DF = Healthcare_df[['Date', 'Close']]
HEALTHCARE_DF = HEALTHCARE_DF.rename(columns={'Close': 'Healthcare Price'})
HEALTHCARE_DF
```

```python
# XLF Financials
path = './CSV_DATA/XLF_FINANCIALS_PRE_2008.csv'
Financials_df = pd.read_csv(path)
FINANCIALS_DF = Financials_df[['Date', 'Close']]
FINANCIALS_DF = FINANCIALS_DF.rename(columns={'Close': 'Financials Price'})
FINANCIALS_DF
```

```python
# XLK Information Technology
path = './CSV_DATA/XLK_INFORMATIONTECHNOLOGY_PRE_2008.csv'
Tech_df = pd.read_csv(path)
TECH_DF = Tech_df[['Date', 'Close']]
TECH_DF = TECH_DF.rename(columns={'Close': 'Information Technology Price'})

TECH_DF
```

```python
# XLU Utilities
path = './CSV_DATA/XLU_UTILITIES_PRE_2008.csv'
Utilities_df = pd.read_csv(path)
UTILITIES_DF = Utilities_df[['Date', 'Close']]
UTILITIES_DF = UTILITIES_DF.rename(columns={'Close': 'Utilities Price'})

UTILITIES_DF
```

```python
# Merge all DataFrames on the 'Date' column
print("Columns of ENERGY_DF:", ENERGY_DF.columns)
print("Columns of MATERIALS_DF:", MATERIALS_DF.columns)
print("Columns of INDUSTRIALS_DF:", INDUSTRIALS_DF.columns)
print("Columns of CONSUMER_DF:", CONSUMER_DF.columns)
print("Columns of STAPLES_DF:", STAPLES_DF.columns)
print("Columns of HEALTHCARE_DF:", HEALTHCARE_DF.columns)
print("Columns of FINANCIALS_DF:", FINANCIALS_DF.columns)
print("Columns of TECH_DF:", TECH_DF.columns)
print("Columns of UTILITIES_DF:", UTILITIES_DF.columns)
```

```python
# Merge all DataFrames on the 'Date' column
merged_df = pd.merge(ENERGY_DF, MATERIALS_DF, on='Date', how='outer')
merged_df = pd.merge(merged_df, INDUSTRIALS_DF, on='Date', how='outer')
merged_df = pd.merge(merged_df, CONSUMER_DF, on='Date', how='outer')
merged_df = pd.merge(merged_df, STAPLES_DF, on='Date', how='outer')
merged_df = pd.merge(merged_df, HEALTHCARE_DF, on='Date', how='outer')
merged_df = pd.merge(merged_df, FINANCIALS_DF, on='Date', how='outer')
merged_df = pd.merge(merged_df, TECH_DF, on='Date', how='outer')
merged_df = pd.merge(merged_df, UTILITIES_DF, on='Date', how='outer')

print(merged_df)
```

```python
merged_df.to_csv('merged_data.csv', index=False)
```
