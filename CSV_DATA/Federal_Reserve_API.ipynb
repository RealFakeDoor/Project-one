```python
Federal_Reserve_Time_Series_Endpoint = 'https://api.stlouisfed.org/fred/series/observations?'
# Series ID:
    # The type of data to return.

# Returns CPI data.
series_id_CPI = 'series_id=CPIAUCNS'

# Returns US 10 year treasury bond yield data.
series_id_Treasury_Bond_Yield = 'series_id=DGS10'

# 10-Year Treasury Constant Maturity Rate: DGS10
# 5-Year Treasury Constant Maturity Rate: DGS5
# 4-Year Treasury Constant Maturity Rate: DGS4
# 3-Year Treasury Constant Maturity Rate: DGS3
# 2-Year Treasury Constant Maturity Rate: DGS2
# 1-Year Treasury Constant Maturity Rate: DGS1

# Returns target federal funding base rates.
series_id_FedRate = 'series_id=FEDFUNDS'

# Determines which output type.
file_type = 'file_type=json'

# limit
# The maximum number of results to return.

# integer between 1 and 1000, optional, default: 1000
```

```python
url = f'{Federal_Reserve_Time_Series_Endpoint}{series_id_FedRate}&{api_key}&{file_type}'

response = requests.get(url).json()
# response

# Convert 'observations' from JSON into DataFrame
json_df = pd.DataFrame(response['observations'])

# Convert 'value' column to numeric
json_df['value'] = pd.to_numeric(json_df['value'])

FEDFUND_DF = json_df.loc[:, ['date', 'value']]
```

```python
# Check the current data type of the 'date' column
print(FEDFUND_DF['date'].dtype)

# Reformat the date column
FEDFUND_DF['date'] = pd.to_datetime(FEDFUND_DF['date'], format='%Y-%d-%m')

# If you want to display the date column in YYYY-MM-DD format
FEDFUND_DF['date'] = FEDFUND_DF['date'].dt.strftime('%Y-%m-%d')

FEDFUND_DF = FEDFUND_DF.rename(columns={'value': 'Federal Fund Rate Target'})

print(FEDFUND_DF)
```

```python
FEDFUND_DF.to_csv('FEDFUND_DF.csv', index=False)
```

```python
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.plot(FEDFUND_DF['date'], FEDFUND_DF['Federal Fund Rate Target'], color='blue', marker='o', linestyle='-')
plt.title('Federal Funding Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Federal Funding Rate')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
